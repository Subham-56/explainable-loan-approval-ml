```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\KIIT0001\OneDrive\Desktop\Project\05_loan_approval\Data\cleaned_loan_data.csv")
```

```python
sns.countplot(x="loan_approved", data=df)
plt.title("Loan Approval Distribution")
plt.show()
```

```python
sns.countplot(x="Credit_History", hue="loan_approved", data=df)
plt.title("Credit History vs Loan Approval")
plt.show()
```

Credit history is the strongest indicator of loan approval.

```python
sns.boxplot(x="loan_approved", y="ApplicantIncome", data=df)
plt.title("Applicant Income vs Loan Approval")
plt.show()
```

Higher applicant income increases approval likelihood.

```python
sns.boxplot(x="loan_approved", y="LoanAmount", data=df)
plt.title("Loan Amount vs Loan Approval")
plt.show()
```

Loan amount alone is not sufficient to determine approval.

```python
sns.countplot(x="Education", hue="loan_approved", data=df)
plt.title("Education vs Loan Approval")
plt.show()
```
